# tf.keras.layers.Dense

* Go through the code and explain values with n_units=1
* Do the same thing with another value of n_units

```
# Initialize a Dense Layer
tf.random.set_seed(1)

n_units = 1

input=np.array([[1., 0., 0.]], dtype=np.float32)
print('input shape:', input.shape)

layer = tf.keras.layers.Dense(n_units)
output = layer(input)
```

```
# Inspect the weights
# There are 2 arrays: one with dimension (n_units,3), another (n_units,)
# These correspond to th kernel and the bias
layer.get_weights()
```

# tf.keras.layers.Conv1D



<img src="https://github.com/rasbt/python-machine-learning-book-3rd-edition/raw/master/ch15/images/15_04.png" width=600>

Reference: Raschka's [ch15_part1.ipynb](https://github.com/rasbt/python-machine-learning-book-3rd-edition/blob/master/ch15/ch15_part1.ipynb)


```
tf.random.set_seed(1)

# Investigate different choices of padding
# Note: there's an error in the figure labels
padding = 'same' # default is "valid"
padding = 'valid'
n_units = 1
kernel_size = 3

# Define the layer
layer = tf.keras.layers.Conv1D(n_units, kernel_size=kernel_size, padding=padding)

# Input dimensions
image_length = 5
n_bands = 3

# padding = 'same' --> output_shape = (image_length, n_units)
# padding = 'valid' --> output_shape = (image_length - 2, n_units)
layer.compute_output_shape(input_shape=(image_length, n_bands))
```

```
# 1 input band
input=np.array([[[1.], [1.], [1.]]], dtype=np.float32)
assert input.shape == (1,3,1)

# 2 input bands
input=np.array([[[1,1], [1,1], [1,1]]], dtype=np.float32)
assert input.shape == (1,3,2)

# initialize the layer with an input
output = layer(input)

batch_size = input.shape[0]
sequence_length = input.shape[1]
n_features = input.shape[2]

# assuming kernel_size = 3
assert kernel_size == 3
if padding == "valid": assert output.shape == (batch_size, sequence_length - 2, n_units)
if padding == "same":  assert output.shape == (batch_size, sequence_length, n_units)


weights = layer.get_weights()
print(weights)

print('\nTrainable parameters:', layer.count_params())
```

```
input=np.array([[[1., 0.], [0., 0.], [0., 0.]]], dtype=np.float32)
assert input.shape == (1,3,2)

layer(input)
```
