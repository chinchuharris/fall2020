# Week 1 -- Introduction

* Artificial Intelligence (AI)
    * Subfield of computer science
    * Theory and development of computer systems that perform tasks normally involving human intelligence
    * Includes computer algorithms that play chess and started beating human masters 40 years ago
    * Includes theory of artificial neurons dating back more than 60 years
* Machine learning (ML)
    * Subfield of AI
    * Self-learning algorithms use data to make predictions
        * ML models have adjustable parameters that are "trained" with data.
        * Trained models use new data to make predictions
    * Statistical foundations (also known as "Statistical Learning") centuries old (Bayes Theorem -- early 1700s)
    * Mathematical foundations (linear algebra, calculus, optimization) dating back to the 1600s
* Deep Learning
    * Subfield of ML
    * Involves networks of millions (or more) artificial neurons
    * Massively parallel calculations requiring modern GPUs
